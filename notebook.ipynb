## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Bagaimana tren rata-rata konsentrasi polutan PM2.5 di stasiun Wanliu secara bulanan selama periode observasi (Maret 2013 - Februari 2017), dan apakah terdapat pola musiman tertentu di bulan-bulan spesifik?
- **Pertanyaan 2:** Bagaimana korelasi (pengaruh) antara parameter cuaca yaitu suhu (TEMP) dan kecepatan angin (WSPM) terhadap tingkat konsentrasi PM2.5 secara keseluruhan di stasiun Wanliu?

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style untuk visualisasi
sns.set(style='whitegrid')
```

## Data Wrangling

### Gathering Data

#### Load df ...

```python
# Memuat dataset Air Quality (Wanliu)
df_wanliu = pd.read_csv("PRSA_Data_Wanliu_20130301-20170228.csv")
df_wanliu.head()
```

### Assessing Data

#### Identifying ... problem

```python
# Melihat informasi umum dataset
print(df_wanliu.info())

# Memeriksa jumlah Missing Values (nilai kosong) di setiap kolom
print(df_wanliu.isnull().sum())
```

**Steps to Take:**
- Missing Value: Terdapat nilai kosong (missing values) pada hampir semua kolom polutan (PM2.5, PM10, SO2, NO2, CO, O3) dan beberapa kolom cuaca.
- Inaccurate Format / Tipe Data: Kolom waktu terpisah-pisah (year, month, day, hour) dalam tipe integer, yang membuat analisis runtun waktu (time-series) menjadi sulit dan tidak efisien.

**Insight:** (Opsional)
- xxx

### Cleaning Data

#### Fixing ... problem

```python
# 1. Fixing Inaccurate Format: Menggabungkan kolom waktu menjadi satu kolom datetime
df_wanliu['datetime'] = pd.to_datetime(df_wanliu[['year', 'month', 'day', 'hour']])

# 2. Fixing Missing Value: Menggunakan metode Forward Fill (ffill)
# Mengingat ini adalah data time-series (deret waktu pengamatan), metode ffill
# sangat cocok karena polusi udara di jam tertentu sangat mirip dengan jam sebelumnya.
df_wanliu.fillna(method='ffill', inplace=True)

# Mengecek kembali apakah data sudah bersih
print("Jumlah missing values setelah dibersihkan:\n", df_wanliu.isnull().sum())
df_wanliu.head()
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Eksplorasi 1: Tren Bulanan Rata-rata PM2.5
# Melakukan resample data berdasarkan bulan (Month)
df_monthly = df_wanliu.resample('M', on='datetime')['PM2.5'].mean().reset_index()

# Eksplorasi 2: Korelasi Parameter Cuaca (TEMP, WSPM) dengan Polusi (PM2.5)
correlation_matrix = df_wanliu[['PM2.5', 'TEMP', 'WSPM']].corr()
print("Matriks Korelasi:\n", correlation_matrix)
```

**Insight:** (Opsional)
- Terdapat fluktuasi rata-rata nilai PM2.5 setiap bulannya.
- Terdapat korelasi negatif antara PM2.5 dengan kecepatan angin (WSPM), yang berarti angin kencang berpotensi membersihkan polusi udara.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_monthly, x='datetime', y='PM2.5', marker='o', color='crimson')

plt.title('Tren Rata-rata Bulanan PM2.5 di Stasiun Wanliu (2013 - 2017)', fontsize=16)
plt.xlabel('Tahun - Bulan', fontsize=12)
plt.ylabel('Rata-rata PM2.5 (µg/m³)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)

plt.title('Heatmap Korelasi PM2.5, Suhu (TEMP), dan Kecepatan Angin (WSPM)', fontsize=14)
plt.show()
```

## Analisis Lanjutan (Opsional)

```python
# Mendefinisikan batas (bins) dan label kategori kualitas udara
# Kriteria standar sederhana: Good, Moderate, Unhealthy, Very Unhealthy, Hazardous
bins = [0, 35, 75, 150, 500, 1000]
labels = ['Good', 'Moderate', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

# Membuat kolom baru menggunakan pd.cut untuk proses binning
df_wanliu['Air_Quality_Category'] = pd.cut(df_wanliu['PM2.5'], bins=bins, labels=labels)

# Visualisasi hasil Binning
plt.figure(figsize=(10, 6))
sns.countplot(data=df_wanliu, x='Air_Quality_Category', palette='viridis')
plt.title('Distribusi Kategori Kualitas Udara (PM2.5) di Stasiun Wanliu', fontsize=14)
plt.xlabel('Kategori Kualitas Udara', fontsize=12)
plt.ylabel('Jumlah Jam Pengamatan', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Conclusion & Recommendation

- **Conclusion pertanyaan 1:** Terdapat pola musiman yang sangat jelas pada tren konsentrasi PM2.5. Tingkat polusi udara (PM2.5) cenderung memuncak atau memburuk secara signifikan pada bulan-bulan musim dingin (akhir tahun hingga awal tahun baru sekitar November - Februari), dan mencapai titik terendah (udara paling bersih) pada bulan-bulan musim panas.

Hal ini membuktikan bahwa tiupan angin yang lebih kencang dapat mengurangi penumpukan polutan di satu tempat. Suhu (TEMP) juga memiliki korelasi negatif namun lebih lemah.
